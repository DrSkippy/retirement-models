```python
import os
import pandas as pd
import string

# Directory containing your CSV files
data_dir = "../output/metrics/"

# Get list of all CSV files in the directory
files = [f for f in os.listdir(data_dir) if f.lower().endswith(".csv")]

dfs = [pd.read_csv(data_dir + fp) for fp in files]

merged_df = pd.concat(dfs)

print(merged_df.describe())
```

```python
import matplotlib
from matplotlib import pylab as plt
```

```python
merged_df.plot.scatter("Period", "net_worth", marker=".", figsize=[12,6])
```

```python
max_period = merged_df.Period.max()
print(max_period)
df_end = merged_df[merged_df["Period"] == max_period]
```

```python
df_end.plot.hist(column=["net_worth"], bins=40)
```
